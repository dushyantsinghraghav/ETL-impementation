```python
unemp_df2.head()
```

```python
unemp_df_new = pd.DataFrame({"Unemployment Rate":unemp_df2.groupby(["State","County"])["Rate"].mean()})
```

```python
unemp_df_new.head()
```

Median Income & Population Dataset

```python
median_inc = median_income_df[['County',"State Code", 'Population',"Median household income"]]
median_inc.head()
```

```python
median_inc = median_inc.rename(columns={"State Code": "State"})
median_inc_new = pd.DataFrame({"Median household income":median_inc.groupby(["State","County"])["Median household income"].mean(),
                              "Population":median_inc.groupby(["State","County"])["Population"].sum()})
```

```python
median_inc_new.head()
```

Now we merge the datasets 

```python
merged = pd.merge(unemp_df_new,median_inc_new, on=["State",'County'],how="left")
```

```python
merged.head()
```

```python
merged2 = pd.merge(merged,new_df2, on=["State",'County'],how="inner")
```

```python
merged.head
```

```python
merged_final = merged2.reset_index() 
```

```python
merged_final.head()
```

```python
merged_final = merged_final[["State","County","Population","Median household income","Unemployment Rate",
                            "Diversity-Index","White","Black","Hispanic/ Latino","Asian","American Indian",
                            "Native Hawaiian","Two or More Races"]]
merged_final.head()
```

```python
merged_final.isna().sum()
#So there are three missing values Median household income
```
